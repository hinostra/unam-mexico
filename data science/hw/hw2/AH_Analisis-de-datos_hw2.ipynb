Acerca del dataset Taylor Swift Spotify

Este dataset contiene informacion de todos los albumes en Spotify de Taylor Swift. 
Los atributos del dataset son
- name: nombre de la cancion
- album: nombre del album
- release_date: año-mes-dia de lanzamiento
- track_number: orden en que la cancion aparece en el album
- id: ID de Spotify de la cancion
- uri: URI de Spotify de la cancion (unique resource indicator code for music on their platform)
- acousticness: medida continua de que tan acustica (not having electrical amplification) es una cancion. Rango de 0.0 a 1.0
- danceability: describe que tan adecuada es una cancion para bailar basado en una combinacion de elementos musicales incluyendo tempo, estabilidad del ritmo, fuerza, etc. Un valor de 0.0 es lo menos bailable y un valor de 1.0 es lo mas bailable.
- energy: medida de 0.0 a 1.0 que representa una medida perceptual de intensidad. Tipicamente las canciones energeticas se sienten rapidas y ruidosas.
- instrumentalness: Rap o palabras habladas se tratan como "vocales". Permite predecir si una cancion contiene o no vocales. Los sonidos "ooh" y "aah" se tratan como instrumentales en este contexto. Mientras mas cerca de 1.0 este valor este, la mayor probabilidad que la cancion no incluya vocales.
- liveness: detecta la presencia de una audiencia en la grabacion. Valores mas altos representan una gran probabilidad que la cancion se haya presentado en vivo.
- loudness: magnitud promedio del sonido percibido en decibeles (dB). La sonoridad es la cualidad de un sonido que constituye el principal correlato psicológico de la intensidad física (amplitud). Los valores suelen oscilar entre -60 y 0 db.
- speechiness: detecta la presencia de palabras habladas en una cancion. Mientras mas cercano sea a 1.0, entonces mayor proporcion de palabras contiene la cancion.
- tempo: tempo (speed or pace of a given piece) promedio de una cancion en beats por minuto (BPM). - The overall estimated tempo of a track in beats per minute (BPM).
- valence: medida de 0.0 a 1.0 que describe la positividad musical que posee una cancion. Canciones con una valencia alta suenan mas positivas (felices, animadas, euforicas), mientras menores valencias indican mas negatividad (tristez, depresion, furia).
- popularity: popularidad de la cancion de 0 a 100.
- duration_ms: duracion de la cancion en milisegundos (ms).

Notamos que muchos albumes tienen multiples versiones, revisemos cuantas versiones tiene cada album.

```python
# nombre de las versiones standard de los albumes
standard_albums = ['Taylor Swift', 'Fearless', 'Speak Now', 'Red', '1989', 'reputation', 'Lover', 'folklore', 'evermore', 'Midnights']
# abreviacion de dichos nombres para los plots
abreviation = ['TS', 'FL', 'SN', 'RED', '1989', 'REP', 'LOV', 'FOL', 'EVE', 'MID']
# colores oficiales de cada album
colors = ['seagreen', 'gold', 'darkorchid', 'darkred', 'skyblue', 'black', 'palevioletred', 'darkgray', 'peru', 'navy']
# escribimos el n_versiones del album Taylor Swift explicitamente para evitar conflictos con otros datos como Taylor Swift Karaoke: reputation
temp_dict = {'standard_name': standard_albums, 'abreviation': abreviation, 'oficial_color': colors, 
             'number_of_versions': [3], 'versions': [['Taylor Swift', 'Taylor Swift (Big Machine Radio Release Special)', 'Taylor Swift (Karaoke Version)']]}

for name in standard_albums[1:]: 
    n = 0               # contador de versiones
    _temp = []          # lista para guardar el nombre de las versiones
    for version_name in data['album'].unique():
        if name in version_name:
            n += 1                          # añadimos 1 al contador cuando el nombre standard forma parte del nombre del album
            _temp.append(version_name)
    temp_dict['number_of_versions'].append(n)
    temp_dict['versions'].append(_temp)

# creamos un dataframe para mayor comodidad
album_data = pd.DataFrame(temp_dict)
# mostramos un barplot
sns.barplot(data=album_data, x='abreviation', y='number_of_versions', palette=album_data.oficial_color)
plt.title('Numero de versiones de cada album en Spotify')
plt.show()
```

```python
# tenemos un nuevo dataframe que nos servira para almacenar informacion de cada album
album_data
```

Una pregunta que podríamos plantearnos sería si a mayor número de versiones, el álbum en cuestión es más popular. Esta hipótesis se basa en el hecho que es lógico suponer que un álbum exitoso presente muchas versiones dado el éxito comercial, es decir, que la disquera produzca múltiples versiones dado el gran consumo del album. Para ello analizaremos el atributo **popularity**. En particular, revisaremos este solo para la versión standard.

```python
# popularidad de cada cancion
for album in album_data.standard_name:
    popularity_songs = data[data.album == album].popularity.tolist()
    print('Indice de popularidad por cada cancion del album', album)
    print(popularity_songs)
```

```python
# resulta mejor ilustrar lo anterior con un box plot (para mejor analisis)
# seguimos limitandonos a las veriones standard
sns.boxplot(data=data, x='popularity', y='album', order=standard_albums, palette=album_data.oficial_color)
plt.show()
```

Notamos que el album con mayor popularidad (Midnights, considerando el valor de la mediana y el rango de popularidad en que se encuentran sus canciones) no es aquel con mayor cantidad de versiones. Lo cual descarta nuestra hipotesis. Ademas de ello, podemos notar que la cancion mas popular tambien pertenece a dicho album y que es un outlier del boxplot. Es interesante notar tambien que los outliers siempre son valores por encima de la mediana, es decir, cada album con outliers parece presentar una o dos canciones mas populares que el resto (que podrian corresponder a los singles principales de dichos albumes los cuales son usualmente los mas populares o escuchados).

Siguiendo con el analisis de popularidad. Entre los albumes incluidos en el dataset se tienen unos que son de presentaciones en vivo. Resulta interesante preguntarnos si las canciones mas populares de los albumes en su version estandar tambien son las canciones mas populares de estas grabaciones de presentaciones en vivo.

```python
# primero seleccionamos aquellas filas con los valores mas altos de popularidad de cada album
# groupby agrupa las filas de data segun el atributo album
# idxmax nos da el indice con maximo valor de popularidad
max_idx = data.groupby('album')['popularity'].idxmax()
max_df = data.loc[max_idx]
max_df
```

Indicamos las canciones mas populares para los siguientes albumes de grabaciones en vivo:
- Live From Clear Channel Stripped 2008: Beautiful Eyes
- Speak Now World Tour Live: Back To December/Apologize/You're Not Sorry
- folklore: the long pond studio sessions: cardigan
- reputation Stadium Tour Surprise Song Playlist: I Don’t Wanna Live Forever (Fifty Shades Darker)

Notamos que a excepcion de *folklore: the long pond studio sessions* (donde la cancion mas popular, cardigan, tambien es la cancion mas popular de su album respectivo), las canciones mas populares de estos albumes de grabaciones en vivo corresponden a canciones no contenidas en este dataset (Beautiful Eyes: parte de un EP no disponible en Spotify, I Don’t Wanna Live Forever: Soundtrack de una pelicula no considerado como parte de la discografia en el dataset) o versiones especiales de otras canciones del album (Back To December/Apologize/You're Not Sorry). Esto parece indicar que los oyentes de estos albumes en vivo parecen preferir las canciones especiales o no contenidas en versiones standard de los albumes, contradiciendo nuestra hipotesis.

Continuamos con el analisis de los otros atributos del dataset. En particular nos interesa el valor **valence**. Mostraremos en una tabla las filas para los valores mas bajos de este atributo en la version standard de los albumes.

```python
subset_data = data[data['album'].isin(album_data.standard_name)]
min_idx = subset_data.groupby('album')['valence'].idxmin()
min_df = subset_data.loc[min_idx]
min_df
```

Notamos que 4 de los albumes en su version standard, presentan el track numero 5 (revisar **track_number**) como el de menor valence o positividad. De hecho es una costumbre de Taylor Swift colocar el track numero 5 como el mas triste/depresivo/sentimental en todos sus albumes. La presencia de resultados diferentes puede deberse a las metricas empleadas por quien genero el dataset, sin embargo resulta interesante notar que el valor que mas se repite es consistente con la vision propia del artista al estructurar el album.

A continuacion, sigamos analizando la positividad de las canciones pero ahora consideremos todas las canciones de un album. 

```python
# para ello haremos uso de un violinplot para analizar la distribucion de los valores de valence en cada album
fig, ax = plt.subplots(figsize=(5,10))
sns.violinplot(data=data, x='valence', y='album', order=standard_albums, palette=colors)
plt.show()
```

En este plot notamos que si bien se tienen canciones con alta positividad en cada album, estas son pocas. Los tres primeros albumes (de genero country) parecen tener una positividad media mientras los albumes con influencias pop (Red, 1989, Lover a excepcion de reputation) parecen presentar mayor positividad (su distribucion esta mas centrada en un valor de valencia mas a la derecha que los primeros 3 albumes). Los siguientes 2 albumes de genero alternativo (folklore, evermore) parecen reducir su nivel de valence, el cual parece ser aun menor en el ultimo album Midnights. Esto parece indicar que segun la metrica del dataset, con cada album, los valores de positividad han ido reduciendose ultimamente. Segun mi perspectiva, esto es debido al cambio en generos musicales y temas tratados en las canciones de la artista.

Por otro lado, analicemos ahora el atributo danceability. Nuevamente aplicaremos esto a las versiones standard de los albumes.

```python
# para ello emplearemos un KDE plot
sns.kdeplot(data=data, x='danceability', hue='album', hue_order=standard_albums, palette=colors)
plt.show()
```

El album Red presenta una distribucion bien centrada (ademas de ser la mas delgada), lo que parece indicar que las canciones de dicho album podrian presentar similares niveles de danceability (lo cual confirmo como fan, ya que hay un balance entre las canciones mas y menos energeticas en dicho album). En general, no notamos que haya algun album que sea predominantemente poco bailable o muy bailable, el ancho de las distribuciones no difiere significativamente entre si (a excepcion del album Red) y parecen cubrir un amplio rango de valores lo que puede indicar que dichos albumes incluyen tanto canciones muy bailables como canciones mas calmadas.

Para culminar con el analisis de los albumes de Taylor Swift, tendremos en cuenta 4 albumes en particular. Desde hace un tiempo, Taylor Swift se propuso regrabar sus primeros 5 albumes para recuperar sus derechos sobre estos. Con esto en mente, lanzo las regrabaciones de los albumes Fearles y Red que ahora llevan por nombre Fearless (Taylor's Version) y Red (Taylor's Version). Compararemos algunos atributos de estos albumes con sus versiones standard (que denominaremos Stolen version) para comparar que tan similares son las versiones antiguas y las nuevas. Consideraremos unicamente canciones incluidas en las versiones standard.

```python
fearless_sv = data[data.album=='Fearless'].loc[600:612, :]
fearless_tv = data[data.album=="Fearless (Taylor's Version)"].loc[63:75, :]
red_sv = data[data.album=='Red']
red_tv = data[data.album=="Red (Taylor's Version)"].loc[33:48, :]
```

```python
# esta vez recurriremos a matplotlib, analizaremos algunos atributos excepto popularity
attributes = [attribute for attribute in data.columns[6:] if attribute!='popularity']

fig, axes = plt.subplots(2, 5, figsize=(5*5, 5.3*2))
for attribute, ax in zip(attributes, axes.flatten()):
    x_0 = pd.concat([fearless_sv[attribute], fearless_tv[attribute], red_sv[attribute], red_tv[attribute]]).reset_index(drop=True).min()
    x_1 = pd.concat([fearless_sv[attribute], fearless_tv[attribute], red_sv[attribute], red_tv[attribute]]).reset_index(drop=True).max()
    ax.plot([x_0, x_1], [x_0, x_1], 'k--', alpha=0.7)
    ax.scatter(x=fearless_sv[attribute], y=fearless_tv[attribute], c=album_data.loc[1, 'oficial_color'], alpha=0.5, label='Fearless')
    ax.scatter(x=red_sv[attribute], y=red_tv[attribute], c=album_data.loc[3, 'oficial_color'], alpha=0.5, label='Red')
    ax.set_title(attribute)
    ax.set_xlabel('Stolen Version')
    ax.set_ylabel("Taylor's Version")
    ax.legend()
plt.show()
```

Para un mejor analisis, trazamos una recta del tipo $x=x$ para ver que tanto difieren algunos atributos de las Stolen Version de las Taylor's Version. El atributo duration_ms es el que presenta mejor ajuste a dicha recta, es decir, aparentemente la duracion de las canciones en ambas versiones es la misma. Sin embargo, en varios atributos se observa una mayor dispersion en torno a la recta, lo que indica un cambio en la produccion o grabacion de las versiones originales y las regrabadas de las canciones. Asi por ejemplo, se encuentra que las canciones de Red y Fearless tienen una mayor danceability para las Stolen Version que para las Taylor's Version (que puede atribuirse a que Taylor Swift era mucho mas joven al grabar las canciones originales por lo que su voz y el estilo de las canciones iba dirigido a un publico mas juvenil).
